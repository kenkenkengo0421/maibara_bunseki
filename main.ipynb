# 凍結防止制御参考
気象庁から取り寄せたCSVデータを参考（１時間ごと）

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("data.csv", encoding="cp932")
df['年月日時'] = pd.to_datetime(df['年月日時'])
```

```python
df
```

```python
plt.figure(figsize=(30, 10))
plt.plot(df['年月日時'], df['気温(℃)'])
plt.axhline(y=0, color='red', linestyle='--')
plt.show()
```

# 1月1日から2月1日の米原市の気温

```python
df_fst = df[(df['年月日時'] >= '2025-01-01') & (df['年月日時'] <= '2025-2-01')]
plt.figure(figsize=(30, 10))
plt.plot(df_fst['年月日時'], df_fst['気温(℃)'])
plt.axhline(y=0, color='red', linestyle='--')
plt.show()
```

# 米原1月1日から2月1日の0時から9時の間

```python
df_fst_am = df_fst[(df_fst['年月日時'].dt.hour >= 0) & (df_fst['年月日時'].dt.hour < 9)]
```
